```python
pokemon.sort_values(['Type 1', 'HP'], ascending = [1, 0])
```

```python
pokemon = pokemon.drop(columns = ['Total'])
```

```python
# Making changes to the data
pokemon['Total'] = pokemon['HP'] + pokemon['Attack'] + pokemon['Defense'] + pokemon['Sp. Atk'] + pokemon['Sp. Def'] + pokemon['Speed']
pokemon.head(5)
```

```python
# Or
pokemon['Total'] = pokemon.iloc[:, 4:10].sum(axis = 1)
pokemon.head(5)
```

```python
# Shifting a Column

# List of Columns
cols = list(pokemon.columns.values)
pokemon = pokemon[cols[0:10] + [cols[12]] + cols[10:12]]
pokemon
```

```python
pokemon.to_csv('Modified.csv', index = False)
# Or
pokemon.to_excel('Modified.xlsx', index = False)

# Valid only for txt files
pokemon.to_csv('Modified.txt', index = False, sep = '\t')
```
